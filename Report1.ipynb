```python
# Import 7 merge data in one CSV file
import pandas as pd
```

```python
df1 = pd.read_csv("D:/Plan/Data Analysis/Intern/Proj2/Report1/delivery_sh.csv")
df2 = pd.read_csv("D:/Plan/Data Analysis/Intern/Proj2/Report1/delivery_yt.csv")
df3 = pd.read_csv("D:/Plan/Data Analysis/Intern/Proj2/Report1/delivery_hz.csv")
df4 = pd.read_csv("D:/Plan/Data Analysis/Intern/Proj2/Report1/delivery_cq.csv")
df5 = pd.read_csv("D:/Plan/Data Analysis/Intern/Proj2/Report1/delivery_delivery_jl.csv")
```

```python
All_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
print(All_df.info())
```

```python
#remove duplicate Value
All_df = All_df.drop_duplicates()
```

```python
print(All_df.value_counts())
```

```python
print(All_df.info())
```

```python
All_df.describe(include = "all")
```

```python
All_df.groupby('city')["accept_gps_lng"].mean()
```

```python
All_df.groupby('city')["accept_gps_lat"].mean()
```

```python
#Assigning the average to null values by cities
```

```python
All_df["accept_gps_lat"] = All_df["accept_gps_lat"].fillna(All_df.groupby('city')["accept_gps_lat"].transform('mean'))
```

```python
All_df["accept_gps_lat"] = All_df["accept_gps_lat"].fillna(All_df["accept_gps_lat"].mean())
```

```python
All_df["accept_gps_lng"] = All_df["accept_gps_lng"].fillna(All_df.groupby('city')["accept_gps_lng"].transform('mean'))
```

```python
All_df["accept_gps_lng"] = All_df["accept_gps_lng"].fillna(All_df["accept_gps_lng"].mean())
```

```python
All_df.describe(include = "all")
```

```python
#Convert object to correct data type time
```

```python
All_df['accept_time'] = pd.to_datetime(All_df['accept_time'], format='%m-%d %H:%M:%S', errors='coerce')
```

```python
All_df['delivery_time'] = pd.to_datetime(All_df['delivery_time'], format='%m-%d %H:%M:%S', errors='coerce')
```

```python
print(All_df.info())
```

```python
#Convert city and aoi_type to numeric
```

```python
All_df = pd.get_dummies(All_df, columns=['aoi_type'], drop_first=True)
```

```python
All_df = pd.get_dummies(All_df, columns=['city'], drop_first=True)
```

```python
# select Latitude range from -90 to 90.
# Longitude range from -180 to 180.
```

```python
All_df = All_df[(All_df['lat'].between(-90, 90)) & (All_df['lng'].between(-180, 180))]
```

```python
# Calculate delivery_duration in minutes
```

```python
All_df['delivery_duration'] = (pd.to_datetime(All_df['delivery_time']) - pd.to_datetime(All_df['accept_time'])).dt.total_seconds() / 60
```

```python
# Detect outliers using Z-score
```

```python
from scipy.stats import zscore
```

```python
# Calculate Z-scores
```

```python
z_scores = zscore(All_df['delivery_duration'].dropna())  # Drop NA for Z-score calculation
outliers = (z_scores < -3) | (z_scores > 3)  # Outlier threshold
All_df = All_df[~outliers]  # Remove outliers
```

```python
print(All_df.isnull().sum())
```

```python
# Save the CSV file
```

```python
All_df.to_csv('All_delivery.csv', index=False)
```
